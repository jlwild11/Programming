```python
import os
import pandas as pd
```

```python
FolderPath = r"N:\GIS\Projects\USFS\Stand_Exams\PHOTO EXPORT\20200724_"
```

```python
ExportFilename = r"C:\Users\jaw\OneDrive - Mountain G. Enterprises, Inc\Desktop\CSV\Filenames.xlsx"
```

```python
Sheetname = "20200724_"
```

```python
fileList = []
for root, dirs, files in os.walk(FolderPath):
    for filename in files:
        fileList.append(filename)
```

```python
#print(fileList)
```

```python
df = pd.DataFrame(fileList, columns=["Filenames"])
df.head()
```

```python
df.to_csv(ExportFilename,index=False)
```
